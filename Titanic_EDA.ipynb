```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('whitegrid')
```

```python
df=pd.read_csv(r"D:\downloads\train.csv")
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
df['Age']=df['Age'].fillna(df['Age'].median())
df['Embarked']=df['Embarked'].fillna(df['Embarked'].mode()[0])
df['Cabin']=df['Cabin'].fillna('Unknown')
```

```python
df.hist(figsize=(12,8)); plt.show()
```

**Observation:** Most passengers are aged 20–40. Fare is right-skewed.

```python
sns.boxplot(x=df['Fare']); plt.show()
```

**Observation:** Fare contains several outliers.

```python
sns.countplot(x='Survived',data=df); plt.show()
```

**Observation:** More passengers died than survived.

```python
sns.countplot(x='Sex',hue='Survived',data=df); plt.show()
```

**Observation:** Females had a higher survival rate.

```python
sns.countplot(x='Pclass',hue='Survived',data=df); plt.show()
```

**Observation:** First-class passengers survived more often.

```python
sns.scatterplot(x='Age',y='Fare',hue='Survived',data=df); plt.show()
```

```python
num=df.select_dtypes(include=np.number)
sns.heatmap(num.corr(),annot=True,cmap='coolwarm'); plt.show()
```
